```python
from python_speech_features import mfcc
from python_speech_features import delta
from python_speech_features import logfbank
import scipy.io.wavfile as wav
import numpy as np
import matplotlib.pyplot as plt


sample_rate, signal = wav.read('OSR_us_000_0010_8k.wav')  # File assumed to be in the same directory
signal = signal[0:int(3.5 * sample_rate)] # check this

# plot the wave
time = np.arange(0,len(signal))*(1.0 / sample_rate)


plt.plot(time,signal,'b')

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Signal in the Time Domain ")
plt.grid('on')
plt.show()


```

```python


mfcc_feat = mfcc(signal,sample_rate,nfilt=40,numcep=12)


#print(mfcc_feat)

plt.pcolormesh(mfcc_feat.T)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

```

```python
d_mfcc_feat = delta(mfcc_feat, 2)

plt.pcolormesh(d_mfcc_feat.T)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
fbank_feat = logfbank(signal,sample_rate)

plt.pcolormesh(fbank_feat.T)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
